```python
path = 'AirPassengers.csv'

df = pd.read_csv(path)

df.head()
```

```python
df.columns =['Date','Number of Passengers']
df.head
```

```python
#matplotlib for data visualization
def plot_df(df, x, y, title=" ", xlabel='Date', ylabel='Number of Passangers',
           dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    
plot_df(df, x=df['Date'], y=df['Number of Passengers'], 
        title='Number of US Airline passangers from 1949 to 1960')
    
```

```python
#Patterns in time-series 
def plot_df(df, x, y, title=' ', xlabel='Date', ylabel='Number of Passangers', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x,y,color='blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    

plot_df(df, x=df['Date'], y=df['Number of Passengers'], title='Trend and Seasonality')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse


# Multiplicative Decomposition 
multiplicative_decomposition = seasonal_decompose(df['Number of Passengers'], model='multiplicative', period=30)


#Plot 

plt.rcParams.update({'figure.figsize':(16,12)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=16)
plt.tight_layout(rect=[0,0.03, 1, 0.95])

plt.show()
```
